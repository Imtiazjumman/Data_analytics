1. Global Fill (df.fillna(0))
df_filled = df.fillna(0)

The Action: This looks at the entire DataFrame (df) and replaces every single missing value (NaN) with 0.

The Result: It creates a new DataFrame called df_filled. The original df remains unchanged.

When to use: This is useful when a missing value logically means "zero" (e.g., a customer who didn't make a purchase has 0 sales).

2. Targeted Imputation (mean())
df["New_Sales"] = df["Sales"].fillna(df["Sales"].mean())

The Action: Instead of filling everything with zero, this focuses only on the "Sales" column.

The Strategy: It calculates the average (mean) of the existing sales numbers and uses that average to plug the holes.

The Result: It creates a new column in your original DataFrame called "New_Sales".

When to use: This is a common statistical technique called "Mean Imputation." It's used when you don't want to bias your data toward zero and prefer to use a "typical" value for that column.

```python
## Change data types
## Renaming Columns

df=df.rename(columns={"Date":"Sales_date"})
df.head()
```

```python
#Change data types
df["New_value"]=df["Value"].fillna(df["Value"].mean()).astype(int)
df.head()
```

The Step-by-Step Logic
.fillna(df["Value"].mean()): It finds any missing (NaN) values in the "Value" column and replaces them with the average of that column.

.astype(int): It converts the entire column from decimals (floats) into whole numbers (integers). This will truncate any decimals (e.g., 10.9 becomes 10).

df["New_value"] = ...: It saves this cleaned, rounded data into a brand new column called "New_value," leaving the original "Value" column untouched.

```python
## Multiply column value apply lambda

df['New Value']= df['Value'].apply(lambda x:x*2).fillna(df['Value'].mean()).astype(int)
df.head()
```

1. Grouping
Grouping is the act of categorizing your data. You aren't changing the values yet; you are just deciding how to "bucket" them.

Example: Grouping a list of employees by their "Department."

2. Aggregating
Aggregation is the math performed on those buckets. It turns a list of many numbers into a single summary number.
Common aggregation functions include:

Sum: Totaling values.

Mean/Average: Finding the typical value.

Count: Seeing how many entries exist in that group.

Min/Max: Finding the range.

```python
## data aggregating and grouping

grouped_mean = df.groupby("Product")["Sales"].mean()
print(grouped_mean)
```

```python
grouped_mean = df.groupby("Product")["New Value"].mean()
print(grouped_mean)
```

```python
#grouped by multiple column
grouped_sum = df.groupby(["Product","Region"])["Value"].sum()
print(grouped_sum)
```

```python
df.groupby(["Product","Region"])["Value"].mean()
```

```python
## Aggregrate Multiple Function
grouped_agg = df.groupby(['Region'])['New Value'].agg(['mean','sum','count'])
print(grouped_agg)
```
