Pandas is a powerful data manipulation library in python, widely used for data analysis and data cleaning. It provides two primary data structures : series and data frame. A series is one dimensional array like object, while data frame is a two dimensional, size mutable and potentially heterogeneous tabular.

```python
import pandas as pd
```

```python
## Series in pandas
import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print("Series:\n",series)
print(type(series))
```

```python
## Create series from dictionary
data = {'a':1,'b':2,'c':3,'d':4,'e':5}
series_dict = pd.Series(data)
print(series_dict)
print(type(series_dict))
```

```python
data = [10,20,30,40,50]
index = ['a','b','c','d','e']
series = pd.Series(data,index)
print(series)
```

```python
#DataFrame
# Create a dataframe from a dictionary of list

data = {
    'Name':['MUHAMMAD','Imtiaz','Ahmed'],
    'Age':[22,24,27],
    'City':['Dhaka','Chittagong','Sylhet']
}

#Now create a data frame
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)  #Convert dict to array
```

```python
#Create  a dataframe from the list of Dictionaries

data = [
    {'Name': 'Muhammad','Age':24,"City":"Dhaka"},
    {"Name":"Ahmed","Age":26,"City":"Chandpur"},
    {"Name":"Imtiaz","Age":28,"City":"Chittagong"}
]
df = pd.DataFrame(data)
print(df)
```
