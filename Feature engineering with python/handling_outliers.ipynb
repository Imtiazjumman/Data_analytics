```python
## Minimum, Maximum, Median, Q1,Q3,IQR
```

```python
import numpy as np
```

```python
list_marks = [41,55,65,47,89,90,99,55,45,100,60,25,70,98,88,78,99,56,70,40,30,95]
```

```python
minimum,maximum,median,Q1,Q3 = np.quantile(list_marks,[0,1.0,0.50,0.25,0.75])
```

```python
print (minimum,maximum,median,Q1,Q3)
```

```python
## IQR এর পূর্ণরূপ হলো Interquartile Range। এটি পরিসংখ্যানের (Statistics) এমন একটি পরিমাপ যা ডেটার মাঝখানের 50% অংশকে নির্দেশ করে। ডেটা কতটা ছড়িয়ে আছে (Spread) এবং ডেটাতে কোনো Outliers (অস্বাভাবিক মান) আছে কি না, তা বোঝার জন্য এটি ব্যবহার করা হয়।
IQR = Q3-Q1
print(IQR)
```

```python
lowe_fence = Q1-1.5*(IQR)
higher_fench = Q3+1.5*(IQR)
```

```python
print('lowe fence',lowe_fence)
print('Higher fench',higher_fench)

## If any value goes under -12.125 and exceed 150.875 its called outliers
```
