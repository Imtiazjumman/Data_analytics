মেশিন লার্নিংয়ে Imbalance Dataset বা ভারসাম্যহীন ডেটাসেট বলতে এমন একটি অবস্থাকে বোঝায়, যেখানে ডেটাসেটের বিভিন্ন ক্লাসের (Categories) নমুনার সংখ্যা সমান নয়। অর্থাৎ, একটি ক্লাসে প্রচুর ডেটা থাকে (যাকে বলা হয় Majority Class), আর অন্য ক্লাসে ডেটার পরিমাণ খুবই কম থাকে (যাকে বলা হয় Minority Class)।

সহজ কথায়, যখন আপনার ডেটাসেটে এক ধরনের তথ্য অনেক বেশি থাকে এবং অন্য ধরনের তথ্য খুব কম থাকে, তখনই তাকে ইমব্যালেন্স ডেটাসেট বলে।

একটি বাস্তব উদাহরণ (Example)
ধরুন, আপনি একটি ব্যাংক লেনদেনের ডেটাসেট নিয়ে কাজ করছেন যেখানে আপনার লক্ষ্য হলো 'Fraud Detection' বা জালিয়াতি ধরা।

মোট ডেটা: ১০,০০০টি লেনদেন।

স্বাভাবিক লেনদেন (Normal): ৯,৯০০টি।

জালিয়াতি লেনদেন (Fraud): ১০০টি।

এখানে স্বাভাবিক লেনদেন হলো ৯৯% এবং জালিয়াতি হলো মাত্র ১%। এটি একটি চরম Imbalance Dataset।

Up-sampling (বা Over-sampling)
এটি হলো কম ডেটা থাকা ক্লাসটির (Minority Class) সংখ্যা বাড়ানো।

কীভাবে কাজ করে: মাইনোরিটি ক্লাসের যে ডেটাগুলো আপনার কাছে আছে, সেগুলোকে বারবার কপি করে বা নতুন কৃত্রিম ডেটা তৈরি করে সংখ্যায় বাড়িয়ে মেজরিটি ক্লাসের সমান করা হয়।

কখন ব্যবহার করবেন: যখন আপনার কাছে মোট ডেটার পরিমাণ খুব কম।
Down-sampling (বা Under-sampling)
এটি হলো বেশি ডেটা থাকা ক্লাসটির (Majority Class) সংখ্যা কমিয়ে ফেলা।

কীভাবে কাজ করে: মেজরিটি ক্লাস থেকে দৈবক্রমে (Randomly) কিছু ডেটা ডিলিট করে দেওয়া হয়, যতক্ষণ না সেটি মাইনোরিটি ক্লাসের সংখ্যার সমান হয়।

কখন ব্যবহার করবেন: যখন আপনার কাছে লক্ষ লক্ষ ডেটা আছে এবং মেজরিটি ক্লাস থেকে কিছু ফেলে দিলেও খুব একটা সমস্যা হবে না।

ধরুন আপনার কাছে ১০০টি 'Fraud' (জালিয়াতি) ট্রানজ্যাকশন আছে এবং ১০,০০০টি 'Normal' ট্রানজ্যাকশন আছে।

Up-sampling করলে: আপনি ১০০টি 'Fraud' ডেটাকে কপি করে বা SMOTE ব্যবহার করে ১০,০০০টি করবেন।

Down-sampling করলে: আপনি ১০,০০০টি 'Normal' ডেটা থেকে ৯,৯০০টি ফেলে দেবেন এবং মাত্র ১০০টি রাখবেন।

```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility

np.random.seed(123)

#Create dataset with two classes

n_samples = 1000 #আপনি আপনার পুরো ডেটাসেটে মোট ১০০০টি তথ্য রাখতে চাচ্ছেন।
class_0_ratio = 0.9 #আপনি চাচ্ছেন যে আপনার প্রথম ক্লাসটি (যাকে আমরা Majority Class বলি) মোট ডেটার ৯০% হোক।
n_class_0 = int(n_samples*class_0_ratio) #১০০০-এর ৯০% বের করা হচ্ছে (1000 \times 0.9 = 900)। অর্থাৎ, 'Class 0'-তে ৯০০টি স্যাম্পল থাকবে।
n_class_1 = n_samples-n_class_0 #বাকি স্যাম্পলগুলো দ্বিতীয় ক্লাসে যাবে (1000 - 900 = 100)। অর্থাৎ, 'Class 1'-এ মাত্র ১০০টি স্যাম্পল থাকবে।
```

```python
n_class_0,n_class_1
```

```python
## Create my dataframe with imbalance dataset

class_0 = pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale = 1,size = n_class_0),
    'feature_2':np.random.normal(loc= 0, scale = 1,size=n_class_0),
    'target': [0]*n_class_0
})

##class_0 তৈরি (Majority Class)এখানে আপনি ৯০০টি (আগের কোড অনুযায়ী n_class_0) স্যাম্পল তৈরি করছেন।feature_1 ও feature_2: এই দুটিরই গড় বা loc=0। এর মানে হলো, এই ক্লাসের বেশিরভাগ ডেটা পয়েন্ট গ্রাফের (0,0) স্থানাঙ্কের আশেপাশে থাকবে।target: সবগুলোর মান দেওয়া হয়েছে 0। এটি মডেলকে বলবে যে, এই বৈশিষ্ট্যগুলো থাকলে সেটি 'Class 0'।

class_1 = pd.DataFrame({
    'feature_1':np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=2,scale = 1, size=n_class_1),
    'target':[1]*n_class_1
    
    ##class_1 তৈরি (Minority Class)এখানে আপনি ১০০টি (আগের কোড অনুযায়ী n_class_1) স্যাম্পল তৈরি করছেন।feature_1 ও feature_2: এই দুটির গড় বা loc=2। এর মানে হলো, এই ক্লাসের ডেটা পয়েন্টগুলো গ্রাফের (2,2) স্থানাঙ্কের আশেপাশে থাকবে।target: সবগুলোর মান দেওয়া হয়েছে 1। এটি মডেলকে বলবে যে, এই বৈশিষ্ট্যগুলো থাকলে সেটি 'Class 1'।##
})
```

```python
df=pd.concat([class_0, class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

```python
##Upsampling
df_minority = df[df['target']==1]
df_majority = df[df['target']==0]
```

scikit-learn-এর এই কোডটি মূলত Oversampling করার জন্য ব্যবহৃত হয়। পয়েন্ট আকারে এর কাজগুলো নিচে দেওয়া হলো:

উদ্দেশ্য: ইমব্যালেন্সড ডাটাবেজে (Imbalanced Dataset) কম সংখ্যক ডাটাকে (Minority) বাড়িয়ে বেশি সংখ্যক ডাটার (Majority) সমান করা।

replace=True: এটি নির্দেশ করে যে একই ডাটা পয়েন্ট একাধিকবার ব্যবহার করা যাবে (বুটস্ট্র্যাপিং)। অর্থাৎ, নতুন স্যাম্পল তৈরির সময় আগের ডাটা বারবার কপি করা হবে।

n_samples=len(df_majority): আপনার মাইনোরিটি ডাটা সেটটি বেড়ে ঠিক কত বড় হবে, তা এটি নির্ধারণ করে। এখানে মেজোরিটি ডাটার সমান সাইজ নির্ধারণ করা হয়েছে।

ফলাফল: কোডটি রান করার পর df_minority এবং df_majority এর ডাটা সংখ্যা সমান হয়ে যায়, যা মডেলকে আরও নির্ভুলভাবে ট্রেন হতে সাহায্য করে।

```python
from sklearn.utils import resample
df_minority_upsample = resample(df_minority, replace=True, n_samples=len(df_majority),
         random_state=42)
```

```python
df_upsampled = pd.concat([df_minority,df_minority_upsample])
```

```python
# Down sampling
import pandas as pd

#Set the random seed for reproducible
np.random.seed(123)

n_samples = 1000 #আপনি আপনার পুরো ডেটাসেটে মোট ১০০০টি তথ্য রাখতে চাচ্ছেন।
class_0_ratio = 0.9 #আপনি চাচ্ছেন যে আপনার প্রথম ক্লাসটি (যাকে আমরা Majority Class বলি) মোট ডেটার ৯০% হোক।
n_class_0 = int(n_samples*class_0_ratio) #১০০০-এর ৯০% বের করা হচ্ছে (1000 \times 0.9 = 900)। অর্থাৎ, 'Class 0'-তে ৯০০টি স্যাম্পল থাকবে।
n_class_1 = n_samples-n_class_0 #বাকি স্যাম্পলগুলো দ্বিতীয় ক্লাসে যাবে (1000 - 900 = 100)। অর্থাৎ, 'Class 1'-এ মাত্র ১০০টি স্যাম্পল থাকবে।

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc = 0,scale = 1,size = n_class_0),
    'feature_2': np.random.normal(loc = 0,scale = 1,size = n_class_0),
    'target': [0]*n_class_0
})

class_1 = pd.DataFrame({
    'feature_1':np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=2,scale = 1, size=n_class_1),
    'target':[1]*n_class_1
})
df = pd.concat([class_0,class_1]).reset_index(drop=True)

#Check the class distribution
print(df['target'].value_counts())
```

```python
#Down sampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==1] 
```

```python
from sklearn.utils import resample
df_majority_downsampled = resample(df_majority,replace = False,
                                   n_samples= len(df_minority),
                                   random_state=42
                                   )
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled = pd.concat([df_minority,df_majority_downsampled])
```
