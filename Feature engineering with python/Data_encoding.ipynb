Data Encoding হলো এমন একটি প্রক্রিয়া যার মাধ্যমে ক্যাটাগরিক্যাল ডেটা (Categorical Data) বা টেক্সট ডেটাকে সংখ্যায় (Numerical values) রূপান্তর করা হয়।মেশিন লার্নিং মডেলগুলো মূলত গাণিতিক সমীকরণের ওপর ভিত্তি করে কাজ করে, তাই তারা সরাসরি "Male/Female" বা "Red/Blue" এর মতো টেক্সট বুঝতে পারে না। মডেলকে ডেটা বোঝানোর জন্য এনকোডিং অপরিহার্য।Data Encoding-এর প্রধান প্রকারভেদপ্রধানত দুই ধরনের এনকোডিং সবচেয়ে বেশি ব্যবহৃত হয়:১. Label Encoding (অর্ডিনাল ডেটার জন্য)যখন ডেটার মধ্যে একটি নির্দিষ্ট ক্রম বা সিরিয়াল (Order) থাকে, তখন এটি ব্যবহার করা হয়। এটি প্রতিটি ক্যাটাগরিকে একটি নির্দিষ্ট নম্বর (০, ১, ২...) দেয়।উদাহরণ: শিক্ষাগত যোগ্যতা (SSC, HSC, BSc)।SSC $\rightarrow$ 0HSC $\rightarrow$ 1BSc $\rightarrow$ 2কখন ব্যবহার করবেন: যখন Rank বা গুরুত্বের ব্যাপার থাকে।২. One-Hot Encoding (নমিনাল ডেটার জন্য)যখন ডেটার মধ্যে কোনো সিরিয়াল বা ক্রম থাকে না, তখন এটি ব্যবহার করা হয়। এটি প্রতিটি ইউনিক ক্যাটাগরির জন্য আলাদা আলাদা নতুন কলাম তৈরি করে এবং সেখানে ০ বা ১ বসায়।উদাহরণ: শহরের নাম (Dhaka, Khulna, Sylhet)।এখানে কেউ কারো চেয়ে বড় বা ছোট নয়।এটি 'Dhaka' এর জন্য একটি কলাম তৈরি করবে যেখানে ঢাকা থাকলে ১ হবে, অন্যথায় ০।
Nominal Data: যেখানে কোনো ক্রম নেই (যেমন: লিঙ্গ, রঙ) One-Hot Encoding ব্যবহার করুন।Ordinal Data: যেখানে ক্রম আছে (যেমন: ভালো, খুব ভালো, চমৎকার) Label Encoding ব্যবহার করুন।

## Nominal/OHE encoding
**One-Hot Encoding**-এর প্রধান সমস্যাগুলো সংক্ষেপে নিচে দেওয়া হলো:

1.  **High Dimensionality:** ক্যাটাগরির সংখ্যা বেশি হলে কলামের সংখ্যাও অনেক বেড়ে যায়। যেমন, ১০০০টি শহরের নামের জন্য ১০০০টি নতুন কলাম তৈরি হবে।
2.  **Sparsity:** নতুন কলামগুলোর অধিকাংশ জায়গা **০ (Zero)** দিয়ে পূর্ণ থাকে, যা মেমরি (RAM) অপচয় করে।
3.  **Dummy Variable Trap:** ভেরিয়েবলগুলোর মধ্যে নিজেদের মধ্যে সম্পর্ক তৈরি হয়, যা লিনিয়ার রিগ্রেশনের মতো মডেলের নির্ভুলতা কমিয়ে দেয়।
4.  **Training Time:** ডেটা বড় হয়ে যাওয়ায় মডেল ট্রেনিং করতে অনেক বেশি সময় লাগে।
5.  **Loss of Order:** ক্যাটাগরির মধ্যে কোনো ক্রম (যেমন: Good > Bad) থাকলে সেই তথ্যটি হারিয়ে যায়।



**সমাধান:** ক্যাটাগরি খুব বেশি হলে **Binary Encoding** বা **Target Encoding** ব্যবহার করা ভালো।

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## Create a simple dataframe
df = pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
## craete an instance of oHE
encoder = OneHotEncoder()

## PErform fit and transform
encoded= encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out()) 
```

```python
encoder_df
```

```python
## For new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis =1)
```

```python
df.head()
```

```python
## Label encoder
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
## For new value
lbl_encoder.transform([['red']])
```

```python
lbl_encoder.transform([['blue']])
```

```python
lbl_encoder.transform([['green']])
```

ORdinal encoding

এনকোডিং এর ধরন,কখন ব্যবহার করবেন?,উদাহরণ
Ordinal Encoding,যখন ডেটার মধ্যে একটি Order বা ক্রম থাকে।,"সন্তুষ্টির মাত্রা (Satisfied, Neutral, Dissatisfied)"
Label Encoding,সাধারণত Target Variable (আউটপুট কলাম) এনকোড করতে এটি ব্যবহার করা হয়।,"ইমেইল ক্যাটাগরি (Spam, Not Spam)"

```python
## Ordinal encoding respect to the ranking
from sklearn.preprocessing import OrdinalEncoder
df = pd.DataFrame({
    'size': ['small','medium','large','small','medium','large']
})
```

```python
df
```

```python
## Asigned rankings
## create an instance of OrdinalEncoder and then fit_transform
encoder= OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```

Target guided ordinal encoding

```python
import pandas as pd
## create a simple datafrmae with a categorical variable and target variable
df = pd.DataFrame({
    'city':['new york','london','paris','tokyo','new york','paris'],
    'price':[200,150,300,250,180,320]
})
df
```

```python
df.groupby('city')['price'].mean()
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded'] = df['city'].map(mean_price)
df[['city', 'price', 'city_encoded']]
```

```python
df[['price', 'city_encoded']]
```
