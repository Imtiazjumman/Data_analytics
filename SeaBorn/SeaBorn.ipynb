```python
import seaborn as sns
```

```python
## Basic plotting with seaborn
tips=sns.load_dataset('tips')
tips
```

```python
#Draw a scatter plot using seaborn
import matplotlib.pyplot as plt

sns.scatterplot(x = 'total_bill',y='tip',data = tips)
plt.title("Scatter plot of Total Bill vs Tips")
plt.show()
```

```python
## Line plot
sns.lineplot(x = 'size', y = 'total_bill', data = tips)
plt.title("Line plot of total bill by size")
plt.show()
```

```python
## categorical Variable plot
## Bar plot
tips.info()
```

```python
sns.barplot(x = "day",y = "total_bill",data = tips)
plt.title("Total bill by day")
plt.show()
```

What does it actually represent?
The Blue Bar: This represents the mean (average) value of all the bills for that specific day.

The Black Line (Error Bar): This represents the 95% Confidence Interval (CI).

Think of it this way: if you went to this restaurant next week, the black line shows the range where the average bill would likely fall.

A long line (like on Friday): This means there is high variability or a smaller sample size; the data is "noisy," and we are less certain about the exact average.

A short line: This means the data points are more consistent, and we are more confident that the average bill really is what the bar shows.

```python
## Box plot
sns.boxplot(x= 'day',y = 'total_bill',data = tips, color = 'Green')
plt.title("Box plot by day and total bill")
plt.show()
```

```python
## Violin plot

sns.violinplot(x= "day",y = "total_bill",data = tips, color = "yellow")
plt.title("Plotting by day and total bill")
plt.show()
```

```python
##Histogram

sns.histplot(tips['total_bill'],bins = 10,kde=True)
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],fill = True,color = 'green')
```

```python
## Pairplot..
sns.pairplot(tips)
```

```python
## Heat map
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot = True,cmap = "coolwarm")
```

```python
import pandas as pd
sales_df = pd.read_csv("data.csv")
display(sales_df.head())
```

```python
# plot total sales by product

plt.figure(figsize=(10,5))
sns.barplot(x = 'Product',y = 'Sales',data = sales_df, estimator=sum)
plt.title('Total sales by product')
plt.xlabel('product')
plt.ylabel('sales')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x = 'Region',y = 'Sales',data = sales_df, estimator=sum)
plt.title('Total sales by product')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```
