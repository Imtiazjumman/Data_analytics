```python
## Feature engineering
##str.split('/'): তারিখটিকে / চিহ্নের জায়গায় ভাগ করে একটি লিস্ট বানায় (যেমন: ["24", "03", "2019"])।
##২. .str[0]: ওই লিস্টের প্রথম উপাদান বা ইনডেক্স ০ (দিন) বেছে নেয়।
##৩. .astype(int): প্রাপ্ত স্ট্রিং ভ্যালুটিকে (যেমন: "24") পূর্ণসংখ্যা বা ইনটেজারে রূপান্তর করে।
##৪. df['Date']: এই রেজাল্টটি Date নামের নতুন একটি কলামে সেভ করে।

df['Date']=df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month']= df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year']= df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
## Since we have extracted the date, month and year from the Date_of_Journey column, we can drop the original Date_of_Journey column as it is no longer needed.
df.drop('Date_of_Journey',axis = 1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Hours']=df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arrival_Minutes']=df['Arrival_Time'].str.split(':').str[1].str.split(' ').str[0].astype(int)
##১. .str.split(':'): সময়টিকে কোলন (:) চিহ্নের জায়গায় ভাগ করে। এর ফলে 18:30 22 Mar হয়ে যায় ["18", "30 22 Mar"]।
##২. .str[1]: লিস্টের দ্বিতীয় অংশটি (index 1) বেছে নেয়, যা হলো "30 22 Mar"।
##৩. .str.split(' '): এবার স্পেস ( ) দিয়ে ভাগ করে যাতে তারিখ আলাদা হয়ে যায়। এটি হয় ["30", "22", "Mar"]।
##৪. .str[0]: এই নতুন লিস্টের প্রথম অংশ অর্থাৎ শুধু মিনিটটি ("30") বেছে নেয়।
## astype(int): প্রাপ্ত স্ট্রিংটিকে পূর্ণসংখ্যা বা ইনটেজারে রূপান্তর করে Arrival_Minutes কলামে সেভ করে।
```
